```python
import pandas as pd
import os
import random
import matplotlib.pyplot as plt
path = "/home/foxcornio/.cache/kagglehub/datasets/mlg-ulb/creditcardfraud/versions/3"
file_path = os.path.join(path, 'creditcard.csv')
data = pd.read_csv(file_path) # Creamos el dataset leyendo el csv

y_real = data['Class'].tolist() # Creamos la lista de datos

print(f"Total de registros cargados: {len(y_real)}") # imprimimos el conteo de y_real
```

```python
y_pred = [] # Almacenara falsos positivos y falsos negativos
for valor in y_real: # Almacenamos los datos mediante un ciclo for 
    if random.random() < 0.95:
        y_pred.append(valor)
    else:
        y_pred.append(1-valor)
data.tail(1)
```

```python
plt.hist(y_real, y_pred)
```
